1. **What is the primary advantage of using convolutional layers in deep learning models?**
   
   A) They reduce the number of parameters compared to fully connected layers.  
   B) They work better with tabular data.  
   C) They are easier to implement than fully connected layers.  
   D) They automatically handle overfitting in image data.

   **Answer**: A) They reduce the number of parameters compared to fully connected layers.

---

2. **In a convolutional neural network (CNN), what does a *filter* or *kernel* do?**
   
   A) It combines features from different layers.  
   B) It performs element-wise multiplication across input data to extract local features.  
   C) It normalizes the output to stabilize learning.  
   D) It shrinks the size of the input data using pooling operations.

   **Answer**: B) It performs element-wise multiplication across input data to extract local features.

---

3. **What is the role of *padding* in convolutional layers?**
   
   A) It increases the size of the input to match the filter size.  
   B) It adjusts the learning rate during training.  
   C) It preserves the spatial dimensions of the input after applying convolutions.  
   D) It normalizes the input data for improved accuracy.

   **Answer**: C) It preserves the spatial dimensions of the input after applying convolutions.

5. **What does the term *stride* refer to in the context of a convolutional neural network (CNN)?**
   
   A) The number of pixels shifted by the filter across the input.  
   B) The size of the filter applied during convolution.  
   C) The number of layers in the network.  
   D) The depth of the convolutional filter.

6. **Why is using multiple filters in each convolutional layer important for a CNN?**
   
   A) It helps reduce overfitting by introducing randomness.  
   B) It allows the network to learn multiple features, such as edges and textures, at different locations.  
   C) It increases the size of the input, improving accuracy.

7. **Which of the following is a key benefit of using *max pooling* in CNNs?**
   
   A) It reduces computational complexity while preserving important features.  
   B) It increases the resolution of the output.  
   C) It prevents overfitting by adding random noise to the input.  
   D) It enhances the sensitivity of the network to minor input variations.

8. **A batch of images with dimensions $(10, 3, 64, 64)$ is passed through a convolutional layer with 32 filters, a filter size of $3 \times 3$, a stride of 1, and no padding. What will be the dimensions of the output batch?**

   PyTorch code:
   ```python
   conv = torch.nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, stride=1, padding=0)
   ```

   A) $(10, 32, 62, 62)$  
   B) $(10, 32, 64, 64)$  
   C) $(10, 32, 60, 60)$  
   D) $(10, 32, 66, 66)$

   **Answer**: A) $(10, 32, 62, 62)$

9. **A batch of 16 images with dimensions $(16, 3, 128, 128)$ is passed through a max pooling layer with a filter size of $2 \times 2$ and a stride of 2. What are the dimensions of the output after pooling?**

   PyTorch code:
   ```python
   pool = torch.nn.MaxPool2d(kernel_size=2, stride=2)
   ```

   A) $(16, 3, 128, 128)$  
   B) $(16, 3, 64, 64)$  
   C) $(16, 3, 126, 126)$  
   D) $(16, 3, 32, 32)$

   **Answer**: B) $(16, 3, 64, 64)$
